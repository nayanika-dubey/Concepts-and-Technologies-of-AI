### Task - 3 - Outlier Detection:

1. Identify outlier countries in South Asia based on their Score and GDP per Capita.
2. Define outliers using the 1.5 × IQR rule.

```python
# Using .quantile() to receive required result and applying formulas
first_quartile = south_asian_happiness_df['score'].quantile(0.25)
second_quartile = south_asian_happiness_df['score'].quantile(0.75)
inter_quartile_range = second_quartile - first_quartile
lower_bound = first_quartile - 1.5 * inter_quartile_range
upper_bound = second_quartile + 1.5 * inter_quartile_range

gdp_first_quartile = south_asian_happiness_df['Log GDP per capita'].quantile(0.25)
gdp_second_quartile = south_asian_happiness_df['Log GDP per capita'].quantile(0.75)
gdp_inter_quartile_range = gdp_second_quartile - gdp_first_quartile
gdp_lower_bound = gdp_first_quartile - 1.5 * gdp_inter_quartile_range
gdp_upper_bound = gdp_second_quartile + 1.5 * gdp_inter_quartile_range

score_outliers = south_asian_happiness_df[(south_asian_happiness_df['score']<lower_bound) | (south_asian_happiness_df['score']>upper_bound)]
gdp_outliers = south_asian_happiness_df[(south_asian_happiness_df['Log GDP per capita']<gdp_lower_bound) | (south_asian_happiness_df['Log GDP per capita']>gdp_upper_bound)]
```

3. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers
in a different color.


```python
# Defining plotting requirements
plt.figure(figsize=(8, 5))
plt.scatter(south_asian_happiness_df['Log GDP per capita'], south_asian_happiness_df['score'], color='green', label='South Asian Countries Score')
plt.scatter(score_outliers['Log GDP per capita'], score_outliers['score'], marker='x', color='Blue', label='South Asian Countries Score Outliers')
plt.scatter(gdp_outliers['Log GDP per capita'], gdp_outliers['score'], marker='+', color='red', label='South Asian Countries GDP Outliers')
plt.title('Outliers in the South Asian GDP per capita and Happiness Score')
plt.xlabel('Log GDP per Capita')
plt.ylabel('Happiness Score')
plt.legend()
plt.show()
```

Outliers are the values that fall at an extreme disproportion from the other values. An outlier can affect the avaerge by **skewing** the value in its favour. In this case, the outlier, when included in finding measurements like mean tend to create a **bais** towards that particular group of data.
